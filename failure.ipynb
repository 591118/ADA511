```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('failure_probability.csv')
df
```

```python
# Example 1: Line Plot
df.plot(x='probability', y='failure_time', kind='line')
plt.title('Line Plot')
plt.xlabel('prob')
plt.ylabel('time')
plt.show()
```

```python
# Example 1: Line Plot with x and y axes switched
df.plot(x='failure_time', y='probability', kind='line')
plt.title('Line Plot')
plt.xlabel('time')
plt.ylabel('prob')
plt.show()
```

```python
# Example 2: Scatter Plot
df.plot(x='probability', y='failure_time', kind='scatter')
plt.title('Scatter Plot')
plt.xlabel('prob')
plt.ylabel('time')
plt.show()
```

```python
# Example 2: Scatter Plot with x and y axes switched
df.plot(x='failure_time', y='probability', kind='scatter')
plt.title('Scatter Plot')
plt.xlabel('time')
plt.ylabel('prob')
plt.show()
```

```python
# Example 3: Bar Plot
df.plot(x='probability', y='failure_time', kind='bar')
plt.title('Bar Plot')
plt.xlabel('prob')
plt.ylabel('time')
plt.show()
```

```python
# Example 3: Bar Plot with x and y axes switched
df.plot(x='failure_time', y='probability', kind='barh')  # Use 'barh' for horizontal bar plot
plt.title('Bar Plot')
plt.xlabel('time')
plt.ylabel('prob')
plt.show()
```

```python
# Example 4: Histogram with x and y axes switched
df['probability'].plot(kind='hist', bins=10, orientation='horizontal')  # Use 'orientation' parameter
plt.title('Histogram')
plt.xlabel('time')
plt.ylabel('prob')
plt.show()
```

```python
# Example 4: Histogram
df['probability'].plot(kind='hist', bins=10)
plt.title('Histogram')
plt.xlabel('prob')
plt.ylabel('time')
plt.show()
```
