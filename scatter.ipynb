```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('scatterXY_A.csv')
df
```

```python
# Using Matplotlib
plt.figure(figsize=(8, 6))  # Optional: Set the figure size
plt.scatter(df['x'], df['y'])
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatterplot of {} vs {}'.format(df['x'], df['y']))
plt.grid(True)
plt.show()

# Using Seaborn
plt.figure(figsize=(8, 6))  # Optional: Set the figure size
sns.scatterplot(data=df, x=df['x'], y=df['y'])
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatterplot of {} vs {}'.format(df['x'], df['y']))
plt.grid(True)
plt.show()
```
