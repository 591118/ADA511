```python
import pandas as pd
```

```python
df = pd.read_csv('income_nominal_data_all.csv')
df
```

```python
df = df[['sex', 'income']]
df
```

```python
count_men_lte_50k = df[(df['sex'] == 'Male') & (df['income'] == '<=50K')].shape[0]
count_men_gt_50k = df[(df['sex'] == 'Male') & (df['income'] == '>50K')].shape[0]
count_women_lte_50k = df[(df['sex'] == 'Female') & (df['income'] == '<=50K')].shape[0]
count_women_gt_50k = df[(df['sex'] == 'Female') & (df['income'] == '>50K')].shape[0]
```

```python
print("Number of men with income <= 50K: ", count_men_lte_50k)
print("Number of men with income > 50K: ", count_men_gt_50k)
print("Number of women with income <= 50K: ", count_women_lte_50k)
print("Number of women with income > 50K: ", count_women_gt_50k)
```

```python
total = df.shape[0]
print(total)
```

```python
percentage_men_lte_50k = (count_men_lte_50k / total) * 100
percentage_men_gt_50k = (count_men_gt_50k / total) * 100
percentage_women_lte_50k = (count_women_lte_50k / total) * 100
percentage_women_gt_50k = (count_women_gt_50k / total) * 100
```

```python
print("Percentage of men with income <= 50K: ", percentage_men_lte_50k)
print("Percentage of men with income > 50K: ", percentage_men_gt_50k)
print("Percentage of women with income <= 50K: ", percentage_women_lte_50k)
print("Percentage of women with income > 50K: ", percentage_women_gt_50k)
```
